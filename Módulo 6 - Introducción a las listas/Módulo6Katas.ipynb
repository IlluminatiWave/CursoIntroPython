## Ejercicio 2: Trabajando con datos de una lista

### Usar slices para recuperar partes de una lista

Es posible que debas trabajar con diferentes secciones de una lista. En nuestro ejemplo, queremos mostrar planetas m치s cerca y m치s lejos del sol de un planeta que el usuario ingresa por teclado.

Comienza agregando el c칩digo para crear una lista con los planetas.

```python
# Lista de planetas
planets = ['Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn', 'Neptune']
```

### Solicita al usuario el nombre de un planeta

A continuaci칩n, agrega el c칩digo para solicitar al usuario un nombre. Debido a que las cadenas distinguen entre may칰sculas y min칰sculas en Python, p칤dale al usuario que use una letra may칰scula para comenzar el nombre del planeta.

```python
# Solicitamos el nombre de un planeta *Pista:  input()*

entrada = input("Introduce el nombre de un planeta: ").title()
print(entrada)

#No es necesario con el .title() 游땙游녧
```

### Encuentra el planeta en la lista
Para determinar qu칠 planetas est치n m치s cerca que el que ingres칩 el usuario, debes encontrar d칩nde est치 el planeta en la lista. Puedes utilizar `index` para realizar esta operaci칩n. Agrega el c칩digo para encontrar el 칤ndice del planeta.

```python
# Busca el planeta en la lista

planets.index(entrada)
```

### Mostrar planetas m치s cercanos al sol que el que el usuario ingres칩

Con el 칤ndice determinado, ahora puedes agregar el c칩digo para mostrar los planetas m치s cercanos al sol.

```python
# Muestra los planetas m치s cercanos al sol

print ("Planetas mas cercanos al Sol que el que escribiste:\n" + str(planets[0:planets.index(entrada)]))
```

# Mostrar planetas m치s alejados del sol que el que el usuario ingres칩

Puedes usar el mismo 칤ndice para mostrar planetas m치s alejados del sol. Sin embargo, recuerda que el 칤ndice inicial se incluye cuando usas un slice. Como resultado, tendr치s que agregar 1 al valor. Agrega el c칩digo para mostrar los planetas m치s alejados del sol.

```python
# Muestra los planetas m치s lejanos al sol

print ("Planetas mas lejanos al Sol que el que escribiste:\n" + str(planets[planets.index(entrada)+1:]))
```

```python
# Solicitamos el nombre de un planeta *Pista:  input()*
user_planet = input('Please enter the name of the planet (with a capital letter to start)')
```

```python
# Muestra los planetas m치s cercanos al sol

print('Here are the planets closer than ' + user_planet)
print(planets[0:planet_index])
```
